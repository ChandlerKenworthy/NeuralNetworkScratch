```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("synthetic_sex_data.csv")
df.head()
```

```python
bins = np.linspace(100.0, 200.0, 50)

plt.figure(figsize=(8,6))
plt.hist(df[df['is_male']==1]['height'], bins=bins)
plt.hist(df[df['is_male']==0]['height'], bins=bins)
plt.show()
```

```python
bins = np.linspace(40.0, 100.0, 100)

plt.figure(figsize=(8,6))
plt.hist(df[df['is_male']==1]['weight'], bins=bins)
plt.hist(df[df['is_male']==0]['weight'], bins=bins)
plt.show()
```
